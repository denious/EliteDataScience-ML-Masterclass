## I. The Search for Empirical Truth

```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('whitegrid')

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('./simulated_data.csv', index_col=0)
df.head()
```

```python
features = df.drop('y', axis=1)
target = df.y
```

```python
def fit_and_plot_model(df, model):
    # Splits input features from target variable
    features = df.drop('y', axis=1)
    target = df.y
    
    # Fits the model
    model.fit(features, target)
    
    # Plots the dataset
    plt.scatter(df.x, df.y)
    
    # Overlays the predicted values for y
    plt.plot(df.x, model.predict(features), 'k--')
    
    # Returns fitted model
    return model


def engineer_polynomial_features(df, max_polynomial=1):
    # Create copy of dataset
    df_copy = df.copy()
    
    # If max_polynomial is 1, then just return the dataframe
    if max_polynomial == 1:
        return df_copy
        
    # Else create polynomial terms for x
    else:
        for power in range(2, max_polynomial + 1):
            df_copy['x{}'.format(power)] = np.power(df_copy.x, power)
        return df_copy

# New fit_and_plot_polynomial_model
def fit_and_plot_polynomial_model(df, model, max_polynomial=1):
    return fit_and_plot_model( engineer_polynomial_features(df, max_polynomial), model )
```

```python
fit_and_plot_polynomial_model(df, LinearRegression(), 3)
```

```python
# mean model
mean_model = np.mean( df.y )
print( mean_model )

# simple linear regression
linear_model = fit_and_plot_model(df, LinearRegression())

# 2nd-order polynomial regression
second_order_polynomial_model = fit_and_plot_polynomial_model(df, LinearRegression(), 2)

# 3rd-order polynomial regression
third_order_polynomial_model = fit_and_plot_polynomial_model(df, LinearRegression(), 3)

# 5th-order polynomial regression
fifth_order_polynomial_model = fit_and_plot_polynomial_model(df, LinearRegression(), 5)

# 10th-order polynomial regression
tenth_order_polynomial_model = fit_and_plot_polynomial_model(df, LinearRegression(), 10)

# unconstrained decision tree
decision_tree_model = fit_and_plot_model(df, DecisionTreeRegressor())

# decision tree, max_depth=3
constrained_tree_by_depth = fit_and_plot_model(df, DecisionTreeRegressor(max_depth = 3))

# decision tree, min_samples_leaf=5
constrained_tree_by_leaf = fit_and_plot_model(df, DecisionTreeRegressor(min_samples_leaf = 5))

# unconstrained random forest
rf = fit_and_plot_model(df, RandomForestRegressor(random_state=123))

# random forest, min_samples_leaf=5
constrained_rf = fit_and_plot_model(df, RandomForestRegressor(min_samples_leaf=5, random_state=123))
```

## II. Unseen Data

```python
# input feature
x_test = np.linspace(0, 2*np.pi, 100)

# noise
np.random.seed(987)
noise_test = np.random.normal(0, .5, 100)

# target variable
y_test = np.sin(x_test) + noise_test
```

```python
test_df = pd.DataFrame({'x': x_test, 'y': y_test})
test_df.head()
```

```python
# Scatterplot of x_test and y_test
plt.scatter(test_df.x, test_df.y)
plt.show()
```

```python
features_test = test_df.drop('y', axis=1)
target_test = test_df.y
```

## III. Performance Metrics

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Predictions on test set from mean model
pred_mean = [mean_model] * len( features_test )

# Predictions on test set from simple linear regression
pred_linear = linear_model.predict( features_test )

# Predictions on test set from 2nd-order polynomial model
pred_second = second_order_polynomial_model.predict(
    engineer_polynomial_features(features_test, 2) )

# Predictions on test set from 3rd-order polynomial model
pred_third = third_order_polynomial_model.predict(
    engineer_polynomial_features(features_test, 3) )

# Predictions on test set from 5th-order polynomial model
pred_fifth = fifth_order_polynomial_model.predict(
    engineer_polynomial_features(features_test, 5) )

# Predictions on test set from 10th-order polynomial model
pred_tenth = tenth_order_polynomial_model.predict(
    engineer_polynomial_features(features_test, 10) )

# Predictions on test set from unconstrained decision tree model
pred_tree = decision_tree_model.predict( features_test )

# Predictions on test set from constrained decision tree by depth
pred_tree_by_depth = constrained_tree_by_depth.predict( features_test )

# Predictions on test set from constrained decision tree by leaf size
pred_tree_by_leaf = constrained_tree_by_leaf.predict( features_test )

# Predictions on test set from random forest
pred_rf = rf.predict( features_test )

# Predictions on test set from constrained random forest
pred_constrained_rf = constrained_rf.predict( features_test )
```

```python
print( 'Mean Squared Error:')
print( np.round(mean_squared_error(target_test, pred_mean), 3), '(Mean Model)' )
print( np.round(mean_squared_error(target_test, pred_linear), 3), '(Linear Model)' )

print( np.round(mean_squared_error(target_test, pred_second), 3), '(2nd-Order Model)' )
print( np.round(mean_squared_error(target_test, pred_third), 3), '(3rd-Order Model)' )
print( np.round(mean_squared_error(target_test, pred_fifth), 3), '(5th-Order Model)' )
print( np.round(mean_squared_error(target_test, pred_tenth), 3), '(10th-Order Model)' )
print( np.round(mean_squared_error(target_test, pred_tree), 3), '(Decision Tree Model)' )

print( np.round(mean_squared_error(target_test,
                                   pred_tree_by_depth), 3), '(Constrained Tree by Depth)' )
print( np.round(mean_squared_error(target_test,
                                   pred_tree_by_leaf), 3), '(Constrained Tree by Leaf)' )

print( np.round(mean_squared_error(target_test, pred_rf), 3), '(Unconstrained Random Forest)' )
print( np.round(mean_squared_error(target_test, pred_constrained_rf), 3), '(Constrained RF)' )
```

```python
print( 'Mean Absolute Error:')
print( np.round(mean_absolute_error(target_test, pred_mean), 3), '(Mean Model)' )
print( np.round(mean_absolute_error(target_test, pred_linear), 3), '(Linear Model)' )

print( np.round(mean_absolute_error(target_test, pred_second), 3), '(2nd-Order Model)' )
print( np.round(mean_absolute_error(target_test, pred_third), 3), '(3rd-Order Model)' )
print( np.round(mean_absolute_error(target_test, pred_fifth), 3), '(5th-Order Model)' )
print( np.round(mean_absolute_error(target_test, pred_tenth), 3), '(10th-Order Model)' )
print( np.round(mean_absolute_error(target_test, pred_tree), 3), '(Decision Tree Model)' )

print( np.round(mean_absolute_error(target_test,
                                    pred_tree_by_depth), 3), '(Constrained Tree by Depth)' )
print( np.round(mean_absolute_error(target_test,
                                    pred_tree_by_leaf), 3), '(Constrained Tree by Leaf)' )

print( np.round(mean_absolute_error(target_test, pred_rf), 3), '(Unconstrained Random Forest)' )
print( np.round(mean_absolute_error(target_test, pred_constrained_rf), 3), '(Constrained RF)' )
```

If you remember from trigonometry, the tangent function is **asymptotic**. As $x$ approaches certain values at regular intervals, then $y$ actually approaches $\infty$ or $-\infty$.

Practically speaking, this means that we will end up with extremely large or extremely small values of $y$ that **appear as outliers** to many machine learning algorithms. Thus, a random forest, which "smoothes out" the predictions, could end up severely punished by predictions near those asymptotes.
